```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_info = pd.read_csv('/content/drive/MyDrive/Datasets/Healthcare insurance fraud/Train-1542865627584.csv')
train_bene = pd.read_csv('/content/drive/MyDrive/Datasets/Healthcare insurance fraud/Train_Beneficiarydata-1542865627584.csv')
train_in = pd.read_csv('/content/drive/MyDrive/Datasets/Healthcare insurance fraud/Train_Inpatientdata-1542865627584.csv')
train_out = pd.read_csv('/content/drive/MyDrive/Datasets/Healthcare insurance fraud/Train_Outpatientdata-1542865627584.csv')
```

```python
train_info
```

```python
train_bene
```

```python
train_in
```

```python
train_out
```

***Hợp nhất các dataframe chứa các bệnh nhân nội trú***

```python
inner_df = pd.merge(train_in, train_bene, on='BeneID', how='inner')
```

```python
final_inner_df = pd.merge(inner_df, train_info, on='Provider', how='inner')
```

```python
final_inner_df
```

```python
final_inner_df.to_csv('Inpatient-Fraud-Dataset.csv', index=False)
```

***Tương tự, làm với các bệnh nhân ngoại trú***

```python
outer_df = pd.merge(train_out, train_bene, on='BeneID', how='inner')
```

```python
final_outer_df = pd.merge(outer_df, train_info, on='Provider', how='inner')
```

```python
final_outer_df
```

```python
final_outer_df.to_csv('Outpatient-Fraud-Dataset.csv', index=False)
```
